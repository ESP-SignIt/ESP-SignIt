# **From Below Run All the Steps to Train and then Download the Trained Model**

```python
NUM_EXAMPLES = 10

for label in labels:
  label_dir = os.path.join(my_folder_path, label)
  example_filenames = os.listdir(label_dir)[:NUM_EXAMPLES]
  fig, axs = plt.subplots(1, NUM_EXAMPLES, figsize=(10,2))
  for i in range(NUM_EXAMPLES):
    axs[i].imshow(plt.imread(os.path.join(label_dir, example_filenames[i])))
    axs[i].get_xaxis().set_visible(False)
    axs[i].get_yaxis().set_visible(False)
  fig.suptitle(f'Showing {NUM_EXAMPLES} examples for {label}')

plt.show()
```

```python
data = gesture_recognizer.Dataset.from_folder(
    dirname=my_folder_path,
    hparams=gesture_recognizer.HandDataPreprocessingParams()
)

print(f"Total data count: {len(data)}")

train_data, rest_data = data.split(0.8)
print(f"Train data count: {len(train_data)}")
print(f"Rest data count: {len(rest_data)}")

validation_data, test_data = rest_data.split(0.5)
print(f"Validation data count: {len(validation_data)}")
print(f"Test data count: {len(test_data)}")

print("Train dataset structure:")
print(train_data)
```

```python
hparams = gesture_recognizer.HParams(
    export_dir="exported_model",
    batch_size=32,
    epochs=10,
    learning_rate=0.001  # Reduced learning rate
)
options = gesture_recognizer.GestureRecognizerOptions(hparams=hparams)

try:
    model = gesture_recognizer.GestureRecognizer.create(
        train_data=train_data,
        validation_data=validation_data,
        options=options
    )
    
    print("Model created successfully")
    print(f"Model architecture: {model.summary()}")

except Exception as e:
    print(f"Error during model creation: {str(e)}")
    import traceback
    traceback.print_exc()
```

```python
loss, acc = model.evaluate(test_data, batch_size=1)
print(f"Test loss:{loss}, Test accuracy:{acc}")
```
